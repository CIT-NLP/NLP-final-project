```python
import re

regex_list = ["drama", "comedy", "action", "adventure", "roman", "crime", "thriller", "family", "world cinema", "horror", "short film", "science fiction", "mystery", "fantasy", "documentary", "silent film"]
for i in range(len(df_3.genres)):     #genre in df_3.genres:
    new_gen_l = []
    for regex in regex_list:
        r = re.compile(regex)
        found = list(filter(r.match, df_3.genres[i]))
        if found:
            new_gen_l.append(regex)
    df_3.genres[i] = new_gen_l
    #print(new_gen_l)
            
        
df_3.genres[1]
```
