```python
import pandas as pd
```

```python
#reading file with genres
columns = ["wikipedia id", "freebase id", "movie name", "year released", "revenue", "runtime", "language", "country released", "genres"]
metadata = pd.read_csv("../Downloads/MovieSummaries/movie.metadata.tsv", sep = "\t", header=None, names = columns)

metadata
```

```python
import json
#genres column is in json format, so load column as dictionary
metadata.genres = metadata.genres.apply(json.loads)
```

```python
#get only genres in each dictionary for each row
metadata.genres = metadata.genres.apply(lambda x : list(x.values()))
```

```python
metadata
```

```python
#reading summary file
columns = ["wikipedia id", "summary"]
summ = pd.read_csv("../Downloads/MovieSummaries/plot_summaries.txt", sep = "\t", header=None, names = columns)

summ
```

```python
#looking at example summary
summ.iloc[1]['summary']
```

```python
#drop unneeded columns
meta = metadata.drop(["freebase id", "year released", "revenue", "runtime", "country released"], axis=1)

#merge dataframes with genres and summaries 
df = pd.merge(meta, summ, how="inner", on="wikipedia id")

df
```

```python
df.language = df.language.apply(json.loads)
df.language = df.language.apply(lambda x : list(x.values()))

df
```

```python
df.language = df.language.apply(lambda x : set(x))

df.language.value_counts()
```

```python
from collections import Counter 

genre_counter = Counter()

df.genres.apply(genre_counter.update)

genre_counter

    
```

```python
common_genres = genre_counter.most_common(35)

common_genres
```
