```python
import pandas as pd
from ast import literal_eval

df = pd.read_csv("./processed_dataframe.csv")

# interpret cols as lists instead of strings
for col in ['unigrams', 'sentences', 'lem_sent', 'lem_sent_join']: 
  df[col] = df[col].apply(literal_eval)

# idk where the col comes from 
df.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
df
```

```python
def make_string(sentences):
  return " ".join(sentences)

df['text'] = df.lem_sent_join.apply(make_string)

df
```

```python
X = []
y = []

for _, row in df.iterrows():
  X.append(row['text'])
  y.append(row['genre_new'])


# y = pd.get_dummies(y).values

y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn import preprocessing
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()

mnb.fit(X_train, y_train)

y_pred = mnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score
print('Testing accuracy %s' % accuracy_score(y_test, y_pred))
print('Testing F1 score: {}'.format(f1_score(y_test, y_pred, average='weighted')))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

conf = confusion_matrix(y_test, y_pred)
classes = list(set(y_train))

f, ax = plt.subplots(figsize=((8,12)))

hmap = sns.heatmap(conf, annot=True, linewidth=.5,fmt='d',ax=ax, square=True, cmap='YlGnBu', vmax=75)
hmap.set_xticklabels(hmap.get_xticklabels(), rotation=45, horizontalalignment='right')
ax.set_xticklabels(classes)
ax.set_yticklabels(classes)

ax.set_xlabel("Predicted Class")
ax.set_ylabel("True Class")
ax.set_title("Confusion Matrix")

```
