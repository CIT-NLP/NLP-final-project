```python
import pandas as pd 
from ast import literal_eval

df = pd.read_csv("./Downloads/data/tokenized.csv")

# interpret cols as lists instead of strings
for col in ['unigrams', 'sentences', 'lem_sent', 'lem_sent_join', 'genres']: 
  df[col] = df[col].apply(literal_eval)

df
```

## Random genre selection

```python
import random

# random genre for single classification
def pick_genre(genres): 
    return random.choice(genres)
```

```python
df_random = df.copy()

df_random['genre'] = df.genres.apply(pick_genre)


df_random
```

```python
df_random.genre.value_counts()
```

```python
df_random['genre'] = df_random.genres.apply(pick_genre)
```

```python
keep_cols = df_random.genre.value_counts().index.tolist()[:7]

df_random = df_random[df_random['genre'].isin(keep_cols)]

df_random
```

```python
df_random.drop('genres', axis = 1, inplace = True)
```

```python
# export dataframe with random genres 
df_random.to_csv('./Downloads/data/random_genre.csv', index = False)
```

## Subsetting genres 

```python
df_subset = df.copy()
```

```python
def get_drama_comedy_action_thriller(genres):
  for gen in genres:
    if gen=="drama":
      return gen
    elif gen=="comedy":
      return gen
    elif gen=="action":
      return gen
    elif gen=="thriller":
      return gen
    else:
      return 0

df_subset["genre"] = df_subset.genres.apply(get_drama_comedy_action_thriller)
```

```python
df_subset
```

```python
# get rid of records not with the 4 chosen genres
df_subset = df_subset[df_subset["genre"] != 0]

df_subset
```

```python
df_subset.genre.value_counts()
```

```python
# subset second to last most samples from each; oversample minority class 

subset_count = df_subset.genre.value_counts().loc['action']

subset_count
```

```python
# take samples from each 

df_drama = df_subset[df_subset["genre"] == "drama"]
df_drama_subset = df_drama.sample(subset_count)

df_comedy = df_subset[df_subset["genre"] == "comedy"]
df_comedy_subset = df_comedy.sample(subset_count)

df_action = df_subset[df_subset["genre"] == "action"]
df_action_subset = df_action.sample(subset_count)

df_thriller = df_subset[df_subset["genre"] == "thriller"]

df_concat = pd.concat([df_drama_subset, df_comedy_subset, df_action_subset, df_thriller])
```

```python
df_concat.genre.value_counts()
```

```python
# prepare for exporting
df_concat.drop('genres', axis = 1, inplace = True)
```

```python
df_concat.to_csv('./Downloads/data/subset_genre.csv', index = False)
```
