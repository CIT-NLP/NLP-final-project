```python
import pandas as pd

df = pd.read_csv("./processed_dataframe.csv")

df
```

```python
# cast columns as lists 
from ast import literal_eval

for col in ['genres', 'unigrams', 'lemmatized', 'sentences', 'lem_sent']: 
  df[col] = df[col].apply(literal_eval)

df
```

```python
df_trunc.genres = df_trunc.genres.apply(lambda gen: gen.split(","))
```

```python
df_trunc
```

```python
def flatten_sent(sent_list):
  ret_val = []
  for sent in sent_list: 
    ret_val.append(" ".join(sent))
  return ret_val 


df_trunc["lem_sent_join"] = df_trunc.lem_sent.apply(flatten_sent)


df_trunc
```

```python
df_trunc.iloc[0]['genres']
```

```python
df_trunc['genre_one'] = df_trunc.genres.apply(lambda gen: gen[0])

df_trunc
```

```python
import string 

def remove_punct(genre): 
  strng= [char for char in genre if char not in string.punctuation]
  return "".join(strng)
    

df_trunc.genre_one = df_trunc.genre_one.apply(remove_punct)
df_trunc.genre_one
```

```python
X = []
y = []


for _, row in df_trunc.iterrows(): 
  X.extend(row['lem_sent_join'])
  for sent in row['lem_sent_join']: 
    y.append(row['genre_one'])

```

```python
X
```

```python
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
mnb = MultinomialNB()

y_pred = mnb.fit(X_train, y_train).predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average = 'weighted') 


print(f"accuracy is {acc}")
print(f"f1 score is {f1}")


```

```python
conf = confusion_matrix(y_test, y_pred)
```
